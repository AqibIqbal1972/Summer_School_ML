```python
import pandas as pd
```

```python
#create a data frame

data = {'Names':['Khan','Faham','Aizaz','Ashir'],
       'Marks':[45,43,30,49]}

pf =pd.DataFrame(data)

pf.plot(x='Names',y='Marks',title='Students Marks')
```

```python
df=pd.DataFrame(data)
df.plot(x='Names',y='Marks',kind='bar',title='Student Marks')
```

```python
df = pd.DataFrame(data)
df.plot(y='Marks', kind='hist', bins=4, title='Students Marks')
```

```python
#Scatter plot
df = pd.DataFrame(data)
df.plot(x='Names', y='Marks', kind='scatter', title='Scatter Plot')
```

```python
#pie chart
df=pd.DataFrame(data)
df.plot(y='Marks', kind='pie', labels=df['Names'], autopct='%1.1f%%', title='Students Marks')
```
