```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('Book1 NaiveBayeTABLe.csv')
data
```

```python
data.info()
```

```python
df = pd.DataFrame(data)
dd = pd.DataFrame(data)
print(df)
```

```python
df.Outlook
```

```python
le = LabelEncoder() # Transform features to numerical values
```

```python
# LabelEncoder in machine learning for converting categorical data into a numerical format that machine learning algorithms can process
df['Outlook'] = le.fit_transform(df['Outlook'])
df['Temperature'] = le.fit_transform(df['Temperature'])
df['Humidity'] = le.fit_transform(df['Humidity'])
df['Windy'] = le.fit_transform(df['Windy'])
df['Class'] = le.fit_transform(df['Class'])
```

```python
df.Outlook
```

```python
df
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
X
```

```python
knn = KNeighborsClassifier(n_neighbors=3)      # use knn algoritham 
knn.fit(X ,y)
```

```python
unseen_sample = [[1, 1, 0, 0]]  # 'rain', 'hot', 'high', 'false‘
predicted_class = knn.predict(unseen_sample)

predicted_class
```

```python
predicted_label = le.inverse_transform(predicted_class)
print("Predicted Class Label:", predicted_label[0])
```
