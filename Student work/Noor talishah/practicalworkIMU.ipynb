```python
import pandas as pd
data ={
     'year':[2013 , 2015 , 2016 , 2017 ,2018, 2020],
      'sales':[50  , 100 ,130  , 200, 180 , 300]}
df = pd.DataFrame( data)     #creat dataFrame
df.plot(x ='year' , y = 'sales' , title = "Sales Over time") 
```

```python
import pandas as pd
# Create a DataFrame
data = {'City': ['karachi', 'lahor', 'Faisalabad', 'Bannu'],
        'Population': [8398748, 5980408, 3716000, 2823006]}
df = pd.DataFrame(data)
# Create a bar plot
df.plot(x='City', y='Population', kind='bar', title='Population by City')

```

```python
import pandas as pd
data = {'Age':[ 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]}
df = pd.DataFrame(data)
# Create a histogram
df.plot(y='Age', kind='hist', bins=5, title='Age Distribution')
```

```python
import pandas as pd
# Create a DataFrame
data = {'X': [3, 2, 1, 7, 5, 2 , 4],
        'Y': [2, 5, 1, 3, 5 , 3 ,5]}
df = pd.DataFrame(data)
# Create a scatter plot
df.plot(x='X', y='Y', kind='scatter', title='Scatter Plot')
```

```python
import pandas as pd
# Create a DataFrame
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [30, 40, 20, 10]}
df = pd.DataFrame(data)
# Create a pie chart
df.plot(y='Value', kind='pie', labels=df['Category'], autopct='%2.1f%%', title='Category Distribution')
```
