```python
import numpy as np
```

```python
arr_1 = np.array([[1,2,3],[4,5,6]])
arr_1
```

```python
                # Sum of all elements in the array
np.sum(arr_1)
```

```python
                # Sum of each row
np.sum(arr_1, axis=1)
```

```python
                # Sum of each column
np.sum(arr_1, axis=0)
```

```python
                # Product of all elements in the array
arr_1.prod()
```

```python
                # Product of each column
arr_1.prod(axis=0)
```

```python
                # Product of each rows
arr_1.prod(axis=1)
```

```python
                # Minimum value in the array
arr_1.min()
```

```python
                # Minimum value in each column
arr_1.min(axis=0)
```

```python
                # Minimum value in each row
arr_1.min(axis=1)
```

```python
                # Maximum value in the array
arr_1.max()
```

```python
                # Index of the minimum element
arr_1.argmin()
```

```python
                # Index of the minimum element in each column
arr_1.argmin(axis=0)
```

```python
                # Index of the maximum element
arr_1.argmin(axis=1)
```

```python
                # Index of the maximum element
arr_1.argmax()
```

```python
                # Mean of the array
arr_1.mean()
```

```python
                # Mean of each column
arr_1.mean(axis=0)
```

```python
                # Standard deviation of the array
arr_1.std()
```

```python
                # Standard deviation of each row
arr_1.std(axis=1)
```

```python
                # Variance of the array
arr_1.var()
```

```python
                # Clip the values of the array between 2 and 4
arr_1.clip(2,4)
```

```python
                # Round the values
arr_2 = np.array([4.12,5.5,6.666])
arr_2.round()
```

```python
                # Round to 2 decimal places
arr_2.round(decimals=2)
```
