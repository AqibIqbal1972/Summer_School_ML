## Scatter Plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
species = iris.target_names

# Create a scatter plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', alpha=0.8)

# Add the legend
handles, _ = scatter.legend_elements()
legend_labels = {i: species[i] for i in range(len(species))}
plt.legend(handles, [legend_labels[i] for i in range(len(species))], title="Species")

plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```

## Line Plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

## Histogram

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(X[:, 3], bins=20, color='purple', alpha=0.7)
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Box Plot


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([X[y==0, 0], X[y==1, 0], X[y==2, 0]], labels=iris.target_names)
plt.title('Sepal Length across Iris Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()
```
